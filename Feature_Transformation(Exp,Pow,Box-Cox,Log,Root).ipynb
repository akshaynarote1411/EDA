```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'x':[10,20,22,19,35,21,15,16,18,15]})
df
```

```python
df["x"].skew()
```

### Transformation :- 
  - It is the process of converting the skewed distribution to normal distribution.
    
#### For Left Skewed Distribution :- 
   - 1. Exponential Transformation
   - 2. Power Transformation
   - 3. Box-Cox Transformation
    
#### For Right Skewed Distribution :- 
   - 1. Log Transformation
   - 2. Root Transformation
   - 3. Box-Cox Transformation

- The above data is skewed distributed data so, we need to convert that data to the normal distribution.
- For converting skewed distribution to normal distribution there are many techniques. 
- So apply that techniques we have to first identify which type of skewed distribution is this.
- The above data distribution is right skewed cause skewness is greater than 1.

### Right Skewed Distribution :-
 #### 1. Log Transformation
 - While applying log first we have to check the minimum value of that column.
 - If the minimum value is 0, then we have to add some neglegible value to that column's value otherwise it will not work because log of 0 is infinite.

```python
df["x"].min()  # Here min value is not 0, so need not add negligilbe value to column(x), if 0 is there in column then we have add negligible value to that column.
```

```python
df["x_log"] = np.log(df["x"])
df
```

```python
df["x_log"].skew()     # It is giving skewness is in [-1,1], so we have converted the skewed data to normal data using Log Transformation
```

#### 2. Root Transformation
- It will apply first square root on every value of column if then also data distribution is not normal then apply further roots(cube root, 4th root, 5th root,...) untill the data distribution is Normally Distributed.  

```python
df["x_root"] = np.sqrt(df["x"])
df
```

```python
df["x_root"].skew()  # It is giving skewness is in [-1,1], so we have converted the skewed data to normal data using Root Transformation
```

#### 3. Box-Cox Transformation
- It covert the all values from column to respected value calculated by the formula (x^λ - 1)/(λ - 1) 
- where lambda is between -5 to 5, except λ=1.
- The best lambda value for the transformation of skewed data to normal distribution is calculated within inside by Box-Cox i.e param return value of boxcox()

```python
from scipy.stats import boxcox

df["x_boxcox"],param = boxcox(df["x"])
print("Labda Value", param)
df
```
