### select_dtypes():

    - df.select_dtypes(include=None, exclude=None) -> Return a subset of the DataFrame's columns based on the column dtypes.
    
    - This function is used to collect the columns from the datasets depending on their datatypes.
    - If we use 'include' parameter then we can get the all the columns of that included datatypes from dataset
    - If we use 'exclude' parameter then we we get the all the columns of other datatypes excluding the excluded datatype from the dataset.

```python
df.select_dtypes(include="object")   # It will gives the data of all the columns which have "object" datatype only.
```

```python
cat = df.select_dtypes(include="object")   # It will gives table(dataframe) of all the categorical columns 
cat
```

```python
df.select_dtypes(include="object").columns   # # It will gives list of all the categorical columns 
```

```python
df.select_dtypes(exclude="object")   # It will gives the data of all the columns which don't have  "object" as datatype(means it will gives all others datatypes columns data execpt "object" datatype.).
```

```python
num = df.select_dtypes(exclude="object")  # It will gives table(dataframe) of all the numerical columns
num
```

```python
df.select_dtypes(exclude="object").columns   # It will gives list of all columns which does not have 'object' as datatype.
```

```python
# Finding count value of each unique category in column


lables = df["continent"].unique()
counts=[]

for i in lables:
    cond = df["continent"]==i
    counts.append(sum(cond))
    
    
counts
```

```python
category = df["continent"].unique()
category
```

```python
d1 = pd.DataFrame(zip(category,counts),columns=(["Continent","counts"]))   # Creating a dataframe of unique categories and their total counts.
d1 
```

#### groupby():

```python
# Combining all three methods for counting unique values from a column

# Method-1 :

lables = df["continent"].unique()
counts=[]

for i in lables:
    cond = df["continent"]==i
    counts.append(sum(cond))
    

print(counts)
print("==================")

# Method-2 :

print(df["continent"].value_counts())
print("==================")

# Method-3 : 

print(df.groupby("continent").size())
```
