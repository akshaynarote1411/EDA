#### Where Condition : 

     - np.where(<condition>,<If_cond_True>,<If_cond_False>)
     - np.where(<Write condition>, <If condition is true,then put that value by which we want to reaplece the old values>, <If condition false, then kept the old value as it is >)

```python
df["Ranks"] = np.where(df["Ranks"]>3,"Participated",df["Ranks"] )   # Here "Participated" is new value with we want to replace the old value if condition is True. Otherwise we have kept the old values as it is idf condition is False
df
```
